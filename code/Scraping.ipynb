```python
#Loading Libraries
import pandas as pd
import requests
```

```python
#Scraping the Data
url = "https://en.wikipedia.org/wiki/List_of_medical_colleges_in_India"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
response = requests.get(url, headers=headers)

tables = pd.read_html(response.content)

# Getting table 3 (the state-wise medical colleges table)
df = tables[3]

print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")


print(df.head(10))
```

```python
# Cleaning the data
# Removing 'Ref.' column
if 'Ref.' in df.columns:
    df = df.drop('Ref.', axis=1)

# Removing the Total row
df = df[df.iloc[:, 0] != 'Total']

# Renaming columns
df.columns = ['State', 'Number_of_Medical_Colleges', 'Number_of_PG_Institutes']

# Converting columns to numeric
df['Number_of_Medical_Colleges'] = pd.to_numeric(df['Number_of_Medical_Colleges'], errors='coerce')
df['Number_of_PG_Institutes'] = pd.to_numeric(df['Number_of_PG_Institutes'], errors='coerce')

# Removing any rows with NaN
df = df.dropna()

print(df)

# Saving the File
df.to_csv('medical_colleges_by_state.csv', index=False)
```
